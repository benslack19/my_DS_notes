## CASE WHEN with aggregate functions

```
-- note the combination of COUNT and using the m.id as a filtering criterion
    
SELECT 
	c.name AS country,
    -- Count games from the 2012/2013 season
	COUNT(CASE WHEN m.season = '2012/2013' 
        	THEN m.id ELSE NULL END) AS matches_2012_2013  
FROM country AS c
LEFT JOIN match AS m    --note use of LEFT JOIN to get all countries
ON c.id = m.country_id
-- Group by country name alias    -- this is taken from tutorial but not sure if you can use alias
GROUP BY country;
```

```
-- use of AVG to calculate percentage following assignment of 1, 0

SELECT 
	c.name AS country,
    -- Calculate the percentage of tied games in each season
	AVG(CASE WHEN m.season='2013/2014' AND m.home_goal = m.away_goal THEN 1
			WHEN m.season='2013/2014' AND m.home_goal != m.away_goal THEN 0
			END) AS ties_2013_2014,
	AVG(CASE WHEN m.season='2014/2015' AND m.home_goal = m.away_goal THEN 1
			WHEN m.season='2014/2015' AND m.home_goal != m.away_goal THEN 0
			END) AS ties_2014_2015
FROM country AS c
LEFT JOIN matches AS m
ON c.id = m.country_id
GROUP BY country;
```

```
-- use of ROUND to limit to 2 decimal places

SELECT 
	c.name AS country,
    -- Round the percentage of tied games to 2 decimal points
	ROUND(AVG(CASE WHEN m.season='2013/2014' AND m.home_goal = m.away_goal THEN 1
			 WHEN m.season='2013/2014' AND m.home_goal != m.away_goal THEN 0
			 END),2) AS pct_ties_2013_2014,
	ROUND(AVG(CASE WHEN m.season='2014/2015' AND m.home_goal = m.away_goal THEN 1
			 WHEN m.season='2014/2015' AND m.home_goal != m.away_goal THEN 0
			 END),2) AS pct_ties_2014_2015
FROM country AS c
LEFT JOIN matches AS m
ON c.id = m.country_id
GROUP BY country;
```

# Simple subqueries

- You can note a subquery since a (SELECT statement) might be contained within parantheses surrounded by another complete SQL statement.
- Useful for intermediary transformations.
- Can be in any part of query (SELECT, FROM WHERE, GROUP BY)
- Can return variety of information (scalar, list, table)

Why subqueries?
- Compare groups to summarized values
- Reshaping data
- Combining data that cannot be joined


What's a simple subquery?
- One that can be run on its own, independent of the outer query
- Only processed once in the entire statement


## Subqueries in WHERE clause

- Useful for filtering results based on information you have to calculate separately, beforehand

```
SELECT date, hometeam_id, awayteam_id, home_goal, away_goal
FROM match
WHERE season = '2012/2013'
      AND home_goal > 
      -- subquery here
      (SELECT AVG(home_goal) FROM match);   
```

- Useful for filtering list with IN

```
SELECT team_long_name,
       team_short_name AS abbr
FROM team
WHERE 
      -- subquery using IN here
      team_api_id IN
      (SELECT hometeam_id
       FROM match
       WHERE country_id = 15722);   
```

# Correlated queries, nested queries, CTEs

## Correlated subqueries

- uses values from the outer query to generate a result
- **re-run for every row generated in the final data set** (this slows things down computationally)
- used for advanced joining, filtering, evalutaing data

- One example I did using all simple subqueries can be made with a correlated subquery


### Simple subquery example

Which match stages tend to have a higher than average number of goals scored?
```
SELECT
    s.stage,
    ROUND(s.avg_goals,2) AS avg_goal,
    (SELECT AVG(home_goal + away_goal)
    FROM match
    WHERE season = '2012/2013') AS overall_avg
FROM (SELECT
    stage,
    AVG(home_goal + away_goal) AS avg_goals FROM match
    WHERE season = '2012/2013'
    GROUP BY stage) AS s -- Subquery in FROM
    
-- note difference between here and below
WHERE s.avg_goals > (SELECT AVG(home_goal + away_goal)
                    FROM match
                    WHERE season = '2012/2013'); -- Subquery in WHERE
```


### Correlated subquery - addressing the same question

```
SELECT
    s.stage,
    ROUND(s.avg_goals,2) AS avg_goal, (SELECT AVG(home_goal + away_goal)
    FROM match
    WHERE season = '2012/2013') AS overall_avg
FROM
    (SELECT stage,
    AVG(home_goal + away_goal) AS avg_goals FROM match
    WHERE season = '2012/2013'
    GROUP BY stage) AS s
    
-- difference between simple subquery and correlated subquery is here
WHERE s.avg_goals > (SELECT AVG(home_goal + away_goal)
                    FROM match AS m
                    WHERE s.stage > m.stage);
--note here's a hallmark of a correlated subquery; using s.stage is dependent on the main query
```



### Simple vs. correlated subqueries


| simple subquery | correlated subquery |
| ------------- | ------ |
| run independently from the main query | dependent on the main query to execute |
| evaluated once in the whole query | evaluated in loops (one for each row), so it's slow |
|  | can slow down performance (don't use too often) |

## Nested subquery

- A query within another query.
- Perform multiple layers of transformation and/or filtering

Example of a nested subquery
```
SELECT
    c.name AS country,
    AVG(m.home_goal + m.away_goal) AS avg_goals,
    AVG(m.home_goal + m.away_goal) -
    
        ---      nested subquery here   ---
        (SELECT AVG(home_goal + away_goal)
         FROM match) AS avg_diff
         
FROM country AS c
LEFT JOIN match as m
ON c.id = m.country_id
GROUP BY country;
```

Nested subqueries can be:
- in multiple layers
- correlated or uncorrelated, or a combination


Example of a nested subqueries, one of which has a correlated subquery

```
SELECT
    c.name AS country,
    
    -- subquery that correlates with main query --
    (SELECT AVG(home_goal + m.away_goal)
     FROM match as m
     WHERE m.country_id = c.id   
           AND id IN 

    -- begin inner subquery   --
               (SELECT id         
                FROM match
                WHERE season = '2011/2012')) AS avg_goals
FROM country AS c
GROUP BY country;
```

## Common Table Expressions (CTE)

Subqueries declared ahead of the main query.

Why use CTEs?
- Executed once
    - CTE is then stored in memory
    - Improves query performance
- Improving organization of queries
- Referencing other CTEs
- Referencing itself (SELF JOIN); a recursive CTE

## Deciding on techniques to use

Different names for the same thing? Joins, subqueries, CTEs. Many can be used interchangeably.
- There's considerable overlap, but not identical

| Joins | Correlated subquery | multiple/nested subqueries |  CTE |
| ------------- | ------ |  ------ | ------ |
| combine 2+ tables | Match subqueries & tables | multi-step transformations |  organize subqueries sequentially |
| simple operations/aggregations| avoid limits of joins (like two columns at once) | improve accuracy and reproducibility |  can reference other CTEs |
|  | high processing time |  |   |



Which to use?
- Depends on database and question
- In general, practice each technique with own databases
- Look for technique that best allows:
    - use and reuse of queries
    - generate clear and accurate results


# Window functions

- It's a way to get around restriction of not being able to use a non-aggregate column with GROUP BY
- Aggregate calculations
    - Similar to subqeries in `SELECT`
    - Running totals, rankings, moving averages
- Processed after every part of query except ORDER BY
- Uses information on already generated result set (a window) rather than database
- Not available in SQLite

```

SELECT 
	-- Select the id, country name, season, home, and away goals
	m.id, 
    c.name AS country, 
    m.season,
	m.home_goal,
	away_goal,
    
    -- Use a window to include the aggregate average in each row
    -- Note how OVER() can immediately precede the AS
    
	AVG(m.home_goal + away_goal) OVER() AS overall_avg
FROM match AS m
LEFT JOIN country AS c ON m.country_id = c.id;
```
