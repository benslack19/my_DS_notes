### One sample example to get sample size for precision [REVISIT]

pwr.p.test(h = , n = , sig.level = power = )

For both two sample and one sample proportion tests, you can specify alternative="two.sided", "less", or "greater" to indicate a two-tailed, or one-tailed test. A two tailed test is the default. 

````r
pwr.p.test(h = 0.5,
           n=NULL,
           sig.level = 0.05,
           power = 0.80,
           alternative = "two.sided")
````

### STAT 509 example

[Comparative treatment efficacy studies example](https://online.stat.psu.edu/stat509/node/43/)

Note that the parameter **d**  in this function is the effect size (Cohen's d) - difference between the means divided by the pooled standard deviation.

**However, the example uses the term "effect size" in a confusing way. I realized I had to use what they called "effect size" and "SD" to get Cohen's d. One way to confirm this is that Cohen's d is unitless.**

````r
d_calc <- 0.25/0.75
````

````r
library(pwr)

pwr.t.test(n=NULL,
           d=d_calc,
           sig.level=.05,
           power=.90,
           alternative="two.sided")
````

# Conditional statements

````r
# filter data
df_batting_100 <- subset(df_batting, PA>100)
````

````r
for (feature in colnames(df_batting_100)) {
    if (typeof(df_batting_100[, feature])=='double') {
        #print(paste(feature, "is double"))
        res <- shapiro.test(df_batting_100[, feature])
        if (res$p.value > 0.01) {
            print(paste(feature, "is normal"))
        }
    }
}
````
