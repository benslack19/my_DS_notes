Make all combinations

Given two arrays and a target value, find a pair of numbers (one from each array) whose sum is closest to the target

Example:
[ 3, 2, 17, -5, 6 ]
[ 5, 4, 12, 0, -81 ]

Target = 13

Return (2, 12)


## Brute force - O(n^2)

```python
def twoSumClosest(nums1, nums2, target):
    '''
    Input: Two arrays, one target
    Output: Pair of numbers, one from each array
    Strategy: Loop through each array, keeping track of sums and distance from target. 
    Keep the pair that results in the closest distance. 
    '''
    import numpy as np
    
    closest_pair = [None, None]
    pair_sums = None
    closest_distance = None
    
    for first_num in nums1:
        for second_num in nums2:
            pair_sums = first_num + second_num
            distance = np.abs(target-pair_sums)
            #print(distance)
            if closest_distance is None:
                closest_distance = distance
                closest_pair = [first_num, second_num]
            elif distance < closest_distance:
                closest_distance = distance
                closest_pair = [first_num, second_num]

    return closest_pair
```

# LC: Palindrome
https://leetcode.com/problems/palindrome-number/


## Brute force

O(n/2)

```python
def is_palindrome(int_value):
    '''
    Is the input value a palindrome?
    '''
    # Turn it into a string to index through
    my_val = str(int_value)
    # See if it's length 1
    if len(my_val)==1:
        return True
    # Iterate but only up to half of the string
    for i in range(round(len(my_val)/2)):
        #print(i)
        if my_val[i] != my_val[-1*(i+1)]:
            return False
    return True
```

## Take first half and just reverse it to see if it matches second half

```python
def is_palindrome_take_substring(int_value):
    '''
    Is the input value a palindrome?
    '''
    # Turn it into a string to index through
    my_val = str(int_value)
    # See if it's length 1
    if len(my_val)==1:
        return True
    # Take first half and just reverse it to see if it matches second half
    half_length = round(len(my_val)/2)
    first_half = my_val[0:half_length]
    second_half = my_val[-half_length:]
    return first_half[::-1]==second_half
```

# LC: Reverse digits of an integer

https://leetcode.com/problems/reverse-integer/

```python
Given a 32-bit signed integer, reverse digits of an integer.

Example 1:

Input: 123
Output: 321
Example 2:

Input: -123
Output: -321
Example 3:

Input: 120
Output: 21
```

## Attempt 1

```python
def reverse(x):
    '''
    x is an integer. return the reverse of it.
    
    Assume we are dealing with an environment which could only 
    store integers within the 32-bit signed integer range: 
    [−231,  231 − 1]. For the purpose of this problem, assume
    that your function returns 0 when the reversed integer
    overflows.
    '''

    x = str(x)
    x = list(x)
    if x[0] != '-':
        x = x[::-1]
        x = ''.join(x)
        x = int(x)
    else:
        x = x[::-1]
        x = x[:-1]
        x = ''.join(x)
        x = int(x)
        x = -1*x
        
    if (x < -2**31) | (x > 2**31-1):
        return 0
        
    return x
```

# LC: Longest Common Prefix (string, array)

Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string "".

Note: All given inputs are in lowercase letters a-z.


## Attempt 1

```python
def longestCommonPrefix(strs):
    '''
    Write a function to find the longest common prefix string amongst an array of strings.
    Input: A list of strings.
    Output: A string, of the longest common prefix string.
    '''
    
    # Brute force
    # Iterate through each string of the list and if 
    # there's something in common, store those letters.
    
    length_of_list = len(strs)
    
    # Iterate through lengths of strings and see if they're equal
    
    prefix_holder = ""
    for subset_length in range(len(strs[0])):
        for i, string in enumerate(strs):
            if subset_length > len(string):
                break
            new_prefix_holder = strs[0][:subset_length]
            if (new_prefix_holder==string[:subset_length]) & (i < length_of_list-1):
                #print('string has common prefix: ')
                continue
            elif (new_prefix_holder==string[:subset_length]) & (i == length_of_list-1):
                #print('string has common prefix and reached end of list')
                prefix_holder = new_prefix_holder

    return prefix_holder
    
```

## Fibonacci

0, 1, 1, 2, 3, 5, 8, 13, 21... 

```python
def fibonacci(n):
    """
    Return the Fibonacci sequence up to the nth number
    """
    
    # Base cases
    if n==0:
        return 0
    elif n==1:
        return 1
    # Solving the problem one step simpler - adding the previous two numbers in the sequence
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

## Low level arrays

- Focus on computer memory
- Memory address
- Units of memory (bits and bytes)
- Memory retrieval

**Low level computer architecture**

- Memory of a computer stored in bits
- Typical units is byte (8 bits)
- Computers typically use a memory address
- Each byte associated with unique address: byte #2144 versus byte #2147
- Individual bytes could have consecutive addresses, the hardware is designed so that any byte of the main memory can be efficiently accessed
- Computer's main memory performs as **random access memory (RAM)**
    - It's as easy to retrieve byte #8672346 as byte #309
    - Indiviudal byte of memory can be stored or retrieved in O(1) time
    
- But an **array** could be stored in consecutive bytes
    - E.g. the 6-character string "SAMPLE" could be stored in 12 consecutive bytes of memory
    - Each letter takes up 16 bites or 2 bytes
    
- Each location within an array is called a **cell**. An integer index describes its location.
    - E.g. The letter "L" could be in the cell of array of index 4 and stored in bytes 2154 and 2155.
- Each cell of an array uses the same number of bytes
    - Allows any cell to be accessed in constant time.
    - Appropriate memory address can be computed using the calculation, **start + (cellsize)(index)**
    - E.g. start of the index for "SAMPLE" might be byte #2146, cellsize is 2, index is 8; that leads to getting byte 2154
    
- This is low level abstraction. But in the real-world, we're talking in the higher level. We just care about indexes 0 to 5 for "SAMPLE" (typical Python index).

**Referential arrays**

- Example: Imagine 100 student names with ID numbers
- Each cell of the array needs to have the same number of bytes. Therefore, we can't use names, since they'd be variable.
- Instead use an array of object  - the **references**
- arr = ['janet', 'bob', 'alice', 'chuck', 'bubba', 'donald', 'sam']
- temp = arr[2:4] - this is a **slice** of the original array
- The temp indexes for 'alice' and 'chuck' are really pointing back to the same locations as arr. Temp isn't a new object, it's **referencing** the original object.
- Most typical copies are shallow copies where most things are just pointing back to the same object.
    - E.g. [0]*8 are all referenceing one existing integer instance
- list_object.extend(extras)  - the extend function is really just adding the reference to the list, it's not really adding to the list
- what does .append do?
    - .extend Iterates over its argument and adding each element to the list and extending the list. The length of the list increases by number of elements in it’s argument. O(n) time.
    - .append Adds its argument as a single element to the end of a list. The length of the list increases by one. O(1) time.

## Using stacks for parentheses problem

https://leetcode.com/problems/valid-parentheses/

```python
def isValid(s):
    '''
    Input: s is a string of just the parentheses characters
    Output: boolean outputting whether the parentheses are balanced
    '''
    
    # If the length of the string is odd, then it can't be balanced
    if len(s)%2 != 0:
        return False
    
    # For each opening bracket, there should be a corresponding closing bracket
    openers = ['(', '{', '[']
    closers = [')', '}', ']']
    parenth_dict = {'(':')', '{':'}', '[':']'}
    
    # If it starts with a closing bracket, it can't be balanced
    if s[0] not in openers:
        return False
    
    # If it ends with an opening bracket, it can't be balanced
    if s[-1] not in closers:
        return False
    
    # Strategy: approach this as a stacks problem (LIFO)
    s = list(s)

    parenth_pair = []
    for i in range(len(s)):
        #print('Before going through. The element: ', s[i], 'parenth_pair', parenth_pair)
        if s[i] in openers:
            parenth_pair.append(s[i])
            #print('add opener: ', parenth_pair)
        if s[i] in closers:
            if len(parenth_pair) == 0:
                return False
            # If the last element in the parenth_pair list matches its closing bracket,
            # then remove that element as it is a legit pair
            elif parenth_dict[parenth_pair[-1]] == s[i]:
                parenth_pair.pop()
                #print('remove opener: ', s[i], 'new parenth_pair_list: ', parenth_pair)
                
    if len(parenth_pair) > 0:
        return False
    else:
        return True
```

### Duplicate removal

Given a string S of lowercase letters, a duplicate removal consists of choosing two adjacent and equal letters, and removing them.

We repeatedly make duplicate removals on S until we no longer can.

Return the final string after all such duplicate removals have been made.  It is guaranteed the answer is unique.

 
Example 1:

Input: "abbaca"
Output: "ca"
Explanation: 
For example, in "abbaca" we could remove "bb" since the letters are adjacent and equal, and this is the only possible move.  The result of this move is that the string is "aaca", of which only "aa" is possible, so the final string is "ca".


```python
def duplicate_removal(s):
    s_nodup = ''
    
    for i in s:
        if (len(s_nodup)==0) & (s_nodup[-1] == i):
            s_nodup = s_nodup[::-1]
            
    return s_nodup
```

## Single number (4/1/20)

Given a non-empty array of integers, every element appears twice except for one. Find that single one.
Note: Your algorithm should have a linear runtime complexity. Could you implement it without using extra memory?

```python
def single_number(nums):
    list_holder = []
    for i in nums:
        """
        Add a new number to the list. Remove it if it's there. Return the remaining number.
        """
        if i not in list_holder:
            list_holder.append(i)
        else:
            list_holder.remove(i)
            
    return list_holder[0]
```

### Strategy 5: Saw solution

```python
def maxSubArray(nums):
    if not nums:
        return 0

    curSum = maxSum = nums[0]
    for num in nums[1:]:
        curSum = max(num, curSum + num)
        maxSum = max(maxSum, curSum)

    return maxSum
```

## Move zeroes (4/4/20)

Given an array nums, write a function to move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Example:

Input: [0,1,0,3,12]
Output: [1,3,12,0,0]
Note:

You must do this in-place without making a copy of the array.
Minimize the total number of operations.

```python
def moveZeroes(nums):
    """
    Do not return anything, modify nums in-place instead.
    """
    cur = 0
    cur_n = 0
    while (cur < len(nums)) & (cur_n < len(nums)):
        if nums[cur]==0:
            nums.pop(cur)
            nums.append(0)
            cur_n += 1
        else:
            cur += 1
        
    return nums
```

### Strategy 1

Three pointers, a left (lower price) and right (to keep checking values) and a third pointer in the middle to be the "memory" of the highest price in that stretch.

Basically, you sell when the price drops again.

Keep track and take a difference.

```python
my_prices = [7,1,5,3,6,4]

cur_L = 0
cur_R = 0
cur_mid = 0
sell_sum = 0

while (cur_R < len(my_prices)) & (cur_mid <= cur_R)  & (cur_L <= cur_mid):
    
    # If the right pointer is at the same place as all of them (like first position)
    # iterate the positions of the mid and R pointers
    if my_prices[cur_R] == my_prices[cur_L]:
        cur_mid += 1
        cur_R += 1
        
    # If the difference between L and mid after iteration goes down,
    # iterate the position of the L pointer
    if my_prices[cur_mid]-my_prices[cur_L] < 0:
        cur_L += 1
        
    # If the difference between L and mid after iteration goes up,
    # hold the midpoint position in "memory" but iterate the right pointer
    if my_prices[cur_mid]-my_prices[cur_L] > 0:
        cur_R += 1
        
    # If the difference between R and mid after iteration goes down,
    # Then move the midpoint position to one less than the right pointer
    # and sell, then reset to the new low position
    if my_prices[cur_R]-my_prices[cur_mid] < 0:
        sell_sum += my_prices[cur_mid]-my_prices[cur_L]
    


print(sell_sum)
```

## Group anagrams (4/6/20)

Group Anagrams
Given an array of strings, group anagrams together.

Example:

Input: ["eat", "tea", "tan", "ate", "nat", "bat"],
Output:
[
  ["ate","eat","tea"],
  ["nat","tan"],
  ["bat"]
]

```python
def groupAnagrams(strs):
    my_dict = {}
    for word in strs:
        sorted_key = "".join(sorted(word))
        if sorted_key in my_dict:
            my_dict[sorted_key].append(word)
        else:
            my_dict[sorted_key] = [word]

    return list(my_dict.values())
```

## Backspace String Compare (4/8/20)

Given two strings S and T, return if they are equal when both are typed into empty text editors. # means a backspace character.

Example 1:

Input: S = "ab#c", T = "ad#c"
Output: true
Explanation: Both S and T become "ac".

```python
def backspaceCompare(S, T):

    mS1 = mT1 = ''
    for i in S:
        if i != '#':
            mS1 += i
        else:
            mS1 = mS1[:-1]

    for i in T:
        if i != '#':
            mT1 += i
        else:
            mT1 = mT1[:-1]
    
    return (mS1==mT1)
```

```python
S = "ab#c"
T = "ad#c" 
backspaceCompare(S, T)
```
